```python
import os
import random
import matplotlib.pyplot as plt

def show_random_cards(cards, rows=1, cols=4):
    # Create a subplot with specified rows and columns
    fig, axes = plt.subplots(rows, cols, figsize=(12, 12))

    # Flatten the axes array if it is multidimensional
    axes = axes.flatten()

    for i in range(rows * cols):
        # Get a random image from the shuffled list
        random_image = random.choice(cards)
        image_path = os.path.join(folder_path, random_image)

        # Display the image using matplotlib on the current subplot
        img = plt.imread(image_path)
        axes[i].imshow(img)
        axes[i].axis('off')  # Turn off axis labels
        axes[i].set_title(f"Card {i+1}")

    plt.show()

# Specify the path to the folder containing your card images
folder_path = "C:\\Users\\win10\Desktop\\DAA CIA-3\\img"

# List all the image files in the folder with specified extensions
card_images = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f)) and f.endswith(('.GIF', '.jpg', '.jpeg'))]

# Show a grid of random cards
show_random_cards(card_images, rows=1, cols=4)
```

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image

folder_path = "C:\\Users\\win10\Desktop\\DAA CIA-3\\img"
file_extension = 'GIF'

# Get a list of all the card files in the folder
card_files = [file for file in os.listdir(folder_path) if file.endswith(f".{file_extension}")]

# Sort the card files to ensure a consistent order
card_files.sort()

# Define the point assignment based on the pattern
points = []
for index, file in enumerate(card_files, start=1):
    if 1 <= index <= 4:  # Assign 1 point to Ace cards (1 to 4)
        points.append(1)
    elif 5 <= index <= 8:  # Assign 10 points to King cards (5 to 8)
        points.append(10)
    elif 9 <= index <= 12:  # Assign 10 points to Queen cards (9 to 12)
        points.append(10)
    elif 13 <=index <=16:
        points.append(10)
    elif 17 <=index <=20:
        points.append(10)
    elif 21 <= index <=24:
        points.append(9)
    elif 25<=index <=28:
        points.append(8)
    elif 29<=index <=32:
        points.append(7)
    elif 33<=index <=36:
        points.append(6)
    elif 37<=index <=40:
        points.append(5)
    elif 41<=index <=44:
        points.append(4)
    elif 45<=index <=48:
        points.append(3)
    elif 49 <=index <=52:
        points.append(2)


```

```python
def calculate_points(cards, points):
    """Calculate the total points of a list of cards."""
    return sum(points[card - 1] for card in cards)

def show_cards(player, cards, points):
    """Display the cards and their total points for a player."""
    print(f"{player}'s cards: {', '.join(map(str, cards))}")
    total_points = calculate_points(cards, points)
    print(f"Total points: {total_points}")
    return total_points

def display_card_images(player, cards):
    """Display the images of the chosen cards for a player."""
    fig, axes = plt.subplots(1, len(cards), figsize=(15, 5))

    for i, card in enumerate(cards):
        image_path = os.path.join(folder_path, f"{card}.GIF")
        img = Image.open(image_path)
        axes[i].imshow(img)
        axes[i].axis('off')  # Turn off axis labels
        axes[i].set_title(f"{player}'s Card {card}")

    plt.show()

def player_turn(player, deck, points):
    """Handle a player's turn."""
    player_hand = []
    
    # Draw 2 initial cards
    player_hand.extend(random.sample(deck, 2))
    deck = [card for card in deck if card not in player_hand]
    
    total_points = show_cards(player, player_hand, points)
    display_card_images(player, player_hand)
    
    while total_points <= 21:
        choice = input(f"{player}, do you want to stay or hit? ").lower()
        
        if choice == 'stay':
            print(f"{player} chose to stay.")
            break
        elif choice == 'hit':
            num_cards_to_draw = int(input(f"{player}, how many cards do you want to draw (1 or 2)? "))
            if num_cards_to_draw in [1, 2]:
                new_cards = random.sample(deck, num_cards_to_draw)
                player_hand.extend(new_cards)
                deck = [card for card in deck if card not in new_cards]
                total_points = show_cards(player, player_hand, points)
                display_card_images(player, player_hand)

                if total_points > 21:
                    print(f"{player} busted! {player} loses.")
                    break
            else:
                print("Invalid choice. Please enter 1 or 2.")
        else:
            print("Invalid choice. Please enter 'stay' or 'hit.'")

    return total_points

# Players' turns
player1_points = player_turn("Player 1", list(range(1, 53)), points)
player2_points = player_turn("Player 2", list(range(1, 53)), points)

# Determine the winner
if player1_points <= 21 and player2_points <= 21:
    if abs(21 - player1_points) < abs(21 - player2_points):
        print("Player 1 wins!")
    elif abs(21 - player2_points) < abs(21 - player1_points):
        print("Player 2 wins!")
    else:
        print("It's a tie!")
elif player1_points <= 21:
    print("Player 1 wins!")
else:
    print("Player 2 wins!")
```
